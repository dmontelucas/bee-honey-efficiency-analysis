```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("../data/processed/honey_pollen_efficiency_clean.csv")
```

```python
X = df[["pollen_input_g", "temperature_c", "humidity_pct", "colony_strength"]]
y = df["honey_output_g"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
model.summary()
```
