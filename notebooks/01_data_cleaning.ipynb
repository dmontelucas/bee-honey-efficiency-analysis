```python
import pandas as pd

data_path = "../data/raw/honey_pollen_efficiency_raw.csv"
df = pd.read_csv(data_path)
```

```python
df = pd.read_csv(data_path)
df.shape
```

```python
df.info()
```

```python
numeric_cols = [
    "collection_1_output",
    "collection_2_output",
    "collection_3_output",
    "collection_4_output",
    "pollen_input_g",
    "honey_output_g"
]

for col in numeric_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .astype(float)
    )

df[numeric_cols].dtypes
```

```python
date_cols = [
    "collection_1_date",
    "collection_2_date",
    "collection_3_date",
    "collection_4_date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], dayfirst=True, errors="coerce")

df[date_cols].head()
```

```python
df["efficiency"] = df["honey_output_g"] / df["pollen_input_g"]
df["honey_per_hour"] = df["honey_output_g"] / df["time_hours"]

df[["pollen_input_g", "honey_output_g", "efficiency", "honey_per_hour"]].head()

```
