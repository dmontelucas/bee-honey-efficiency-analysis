# Efficiency Analysis: Summary and Scope

This notebook summarises findings from the data cleaning, exploratory analysis,
and database aggregation stages. The goal is to interpret efficiency trends,
identify limitations, and propose next analytical steps. Data analysed has been taken as part of a study to establish honey production efficiency from varied pollen input amounts. 


# Dataset Overview

- Number of trials: ~107
- Key variables analysed:
  - Input quantity (`pollen_input_g`)
  - Output quantity (`honey_output_g`)
  - Derived efficiency metric
- Data processed and validated in Notebook 1
- Patterns explored via EDA (Notebook 2) and SQL aggregation (Notebook 4)


# Key Findings and Intepretation
- trial data demonstrates that highest averages / peak efficiency occurred around 2000g (661.6 %) of supplied pollen to experimental hives of varying populations (80,000 - 160,000). Post this 2000g mark, trial data indicates a downward sloping trend where more pollen does not yield more honey produced (3000g, 556.1%).

# Limitations
However, these conclusions are limited by number of varying trials. 94 trials were conducted with 2000g input of pollen, weres considerably less were done at other intervals with 2 operating at 0-499g input, 1 at 500-1499g, 3 at 1500-1999g, and 7 at 3000g. conclusions drawn about efficiency at 2000g can be considered highly reliable with this not being applicable to conclusions drawn from data related to all other input buckets of the study. 

# Recommendations & Next Steps
Due to ongoing data reporting for, it is important in the short-term to collect additional trial data for underrepresented ranges to better validate efficiency stability across ranges.

In the longer term, the findings of this study can be expanded by subjecting trial data to further analysis via regression modelling, non-linear modelling, and predictive modelling. These analyses should be performed to better quantify input-output relationships, detect saturation points, and estimate optimal input levels respectively.

# Final Summary
This analysis identifies a clear efficiency trend that appears to peak around 2000g of pollen input despite trial data limitations. These results justify the need for further targeted data collection and deeper data analysis.
